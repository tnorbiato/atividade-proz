```python
quantRodas = int(input("Digite a quantidade de rodas: "))
pesoVeiculo = int(input("Digite o peso bruto do veículo em quilogramas: "))
quantPessoas = int(input("Digite a quantidade de pessoas no veículo: "))

if (quantRodas < 4) and (quantRodas >= 2):
    print("A melhor categoria de habilitação é a A.")
elif (quantRodas == 4) and (pesoVeiculo <= 3500) and (quantPessoas <= 8):
    print("A melhor categoria de habilitação é a B.")
elif (quantRodas >= 4) and (pesoVeiculo >= 3500 and pesoVeiculo <= 6000):
    print("A melhor categoria de habilitação é a C.")
elif (quantRodas >= 4) and (quantPessoas > 8):
    print("A melhor categoria de habilitação é a D.")
elif (quantRodas >= 4) and (pesoVeiculo > 6000):
    print("A melhor categoria de habilitação é a E.")
```
